Get a list of non-vegan ingredients. This definitely isn't the most clean list but it'll have to do for now.

```python
nonVeganIngredients = set()
for base_ingred in base_non_vegan:
    for ingred in ingred_count:
        if base_ingred in ingred:
            nonVeganIngredients.add(ingred)
nonVeganIngredients.update(manual_vegan_ingredients)
```

Create a new dataframe with the schema {recipe_id: int, isVegan: bool}

```python
def isVegan(ingredients, recipe_name, nonVeganIngredients):
    for ingredient in ingredients:
        if ingredient in nonVeganIngredients:
            return False
        
    for word in recipe_name.split():
        if word in nonVeganIngredients:
            return False
    
    return True
```

```python
isVeganDF = []
for i in range(df.shape[0]):
    recipe_name = df.iloc[i].recipe_name.lower()
    ingredients = df.iloc[i].ingredients.lower().split("^")
    
    vegan = isVegan(ingredients, recipe_name, nonVeganIngredients)
    recipe_id = df.iloc[i].recipe_id
    
    isVeganDF.append([recipe_id, vegan])
    
isVeganDF = pd.DataFrame(isVeganDF, columns=["recipe_id", "isVegan"])
```
